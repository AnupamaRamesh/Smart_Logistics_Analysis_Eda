INTRODUCTION

The Smart Logistics Dataset provides a comprehensive view of logistics operations, capturing key parameters that influence transportation efficiency and supply chain performance. The dataset contains 1,001 records with details such as shipment status, inventory levels, environmental conditions (temperature, humidity), traffic status, waiting times, and user transaction behaviors.

This project aims to perform Exploratory Data Analysis (EDA) to uncover patterns, trends, and insights that impact logistics delays and asset utilization. By analyzing factors like traffic conditions, weather, and user demand, the study seeks to identify major causes of delivery delays and propose data-driven recommendations for optimizing logistics efficiency.

The dataset’s diverse attributes — ranging from asset tracking (Latitude, Longitude) to operational metrics (Inventory Level, Asset Utilization, Demand Forecast) — make it ideal for understanding how multiple variables interact to influence overall supply chain performance.

What is the frequency of shipments under different traffic conditions?

What is the average waiting time for shipments under different traffic conditions ?

```python
traffic_waitingtime = df.groupby("Traffic_Status")["Waiting_Time"].mean()#average waiting time by different traffic condition
print(traffic_waitingtime)

import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
traffic_waitingtime.plot(kind="bar", color=["green", "yellow", "red"])
plt.xlabel("Traffic Condition")
plt.ylabel("Average Waiting Time")
plt.title("Average Waiting Time by Traffic Condition")
plt.show()
```

The bar chart shows that the average waiting time is highest during Clear traffic (≈35.54 minutes), slightly lower during Detour (≈35.31 minutes), and lowest during Heavy traffic (≈34.32 minutes). This suggests that waiting time is relatively consistent across all traffic types, with minimal variation, indicating that factors other than traffic condition might play a larger role in determining shipment waiting times.

What are the most common logistics delay reasons affecting shipment deliveries ?

```python
#Distribution of Logistics Delay Reasons
delay_reason = df["Logistics_Delay_Reason"].value_counts()
print(delay_reason)

#logistics delay reasons
import seaborn as sns
plt.figure(figsize=(8, 5))
delay_reason.plot(kind="bar", color=["orange", "red", "violet"])
plt.xlabel("Logistics Delay Reason")
plt.ylabel("Count of Delayed Shipments")
plt.title("Distribution of Logistics Delay Reasons")
plt.show()


```

Which assets have the highest average utilization rates in the logistics fleet?

What is the relationship between demand forecast and user behavior variables such as transaction amount and purchase frequency?

```python
#how does purchase behaviour impact forecast demand
#showing relationship b/w User_Transaction_Amount andDemand_Forecast
Correlation_value_Transaction=df['User_Transaction_Amount'].corr(df['Demand_Forecast'])
print("showing relationship between 'User_Transaction_Amount and 'Demand_Forecast'",Correlation_value_Transaction)
#showing relationship b/w User_purchase_Fequency and Demand_forecast
Correlation_value_Purcahse=df['User_Purchase_Frequency'].corr(df['Demand_Forecast'])
print("showing relationship between 'User_Purchase_Frequency' and 'Demand_Forecast'",Correlation_value_Purcahse)
```

The correlation results show a very weak positive relationship (0.0045) between User_Transaction_Amount and Demand_Forecast, and a slightly negative relationship (-0.0829) between User_Purchase_Frequency and Demand_Forecast. This indicates that neither transaction amount nor purchase frequency has a strong influence on the demand forecast, suggesting that other factors may play a more significant role in determining demand patterns.

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,5))
sns.barplot(data=df,x='User_Transaction_Amount',y='Demand_Forecast',hue='User_Transaction_Amount',palette="husl",legend=False)
plt.title("Average Demand Forecast by User Transaction Amount")
plt.xlabel("User Transaction Amount")
plt.ylabel("Average Demand Forecast")
plt.tight_layout()
plt.show()
```
